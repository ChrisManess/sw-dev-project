Plan is to use the API to get a list of available newspapers the get the data for a couple

```python
import requests
import pprint

pp = pprint.PrettyPrinter(indent=4)

#all these methods are inspired by the example code. Just wanted to recognize that.
def get_list_of_available_newspapers():
    url = f"https://chroniclingamerica.loc.gov/newspapers.json"
    rsp = requests.get(url)
    if rsp.status_code != 200:
        raise Exception(f"Unable to load {url}")
    newspaper_data = rsp.json()
    return newspaper_data['newspapers']
```

```python
list_of_available_newspapers = get_list_of_available_newspapers()
print(len(list_of_available_newspapers))
pp.pprint(list_of_available_newspapers[0])
```

Awesome that's a lot of newspapers though definitely don't want to go through all them (it's outside the scope of this and I'm a little worried about running into a rate limit. I've had sites ban my IP for scraping it's a fun story you should ask me about it) so let's just use the first 100 and write a function to get their information from the url provided.

```python
def get_info_for_newspaper(id=None):
    url = f"https://chroniclingamerica.loc.gov/lccn/{id}.json"
    rsp = requests.get(url)
    if rsp.status_code != 200:
        raise Exception(f"Unable to load {url}")
    newspaper_data = rsp.json()
    return newspaper_data

pp.pprint(get_info_for_newspaper(id="sn86072192"))
```

Wow that's a lot of information we get this will help us in determining which newspapers to pull down. Let's grab the first 100.

```python
newspaper_subset = list_of_available_newspapers[:100] #taking a sort of immutable object approach here. We don't want to modify the original
papers = []
for paper in newspaper_subset:
    papers.append(get_info_for_newspaper(id=paper['lccn']))
```

```python
papers_with_more_than_99_issues = list(filter(lambda x: len(x['issues']) > 99 , papers))
len(papers_with_more_than_99_issues)
```

Ok that looks better to me. Now we need to retreive the OCR contents of the first 100 issues of each. Once again we need to be thinking about API limits here.
```python
#stealing this from the example
def get_issue(_id):
    if _id.startswith("https"):
        url = _id
    else:
        url = f"https://chroniclingamerica.loc.gov/lccn/{_id}.json"
    doc = {
        "id": _id,
        "source": url
    }
    rsp = requests.get(url)
    if rsp.status_code != 200:
        raise Exception(f"Unable to load {url}")
    issue_data = rsp.json()
    return issue_data

get_issue(paper_0['issues'][0]['url'])
```

From what I can tell about the API each sequence is a page and has it's own OCR url. For example (https://chroniclingamerica.loc.gov/lccn/sn86069873/1900-01-05/ed-1/seq-1/ocr.txt) points to the text output for a particular file.

Let's slow down here for a minute and examine that output. The OCR quality is pretty rough so we are left with some jumbled text. That's really going to limit what we can do in the data analysis phase that would produce useful results. I was hoping for some sentiment analysis or something similarly fun but as they say garbage in garbage out. So we'll end up doing something simpler like word count, maybe tf-idf but with the amount of junk data it wouldn't do well.

So now we have to write a function to retreive the issues and their pages and grab their data to store them on the file system. First lets see how many issues each of these papers has.

```python
print(len(paper_0['issues']))
print(len(paper_1['issues']))
```

Glad we did that check we're only going to take the first 100 from both

```python
paper_0_issues_to_retrieve = paper_0['issues'][:100]
paper_1_issues_to_retrieve = paper_1['issues'][:100]
```

Now we're down to the last bit. Retreival and transforming to be in this format:
`{
    "source": "https://chroniclingamerica.loc.gov/lccn/sn98060264/1917-11-20/ed-1.json",
    "title": "Valdez daily prospector 1917-11-20",
    "isPartOf": "Valdez daily prospector",
    "datePublished": "1917-11-20",
    "volume": "14",
    "number": "49",
    "text": [
        "page 1.....",
        "page 2.....",
        "page 3.....",
        "page 4....."
    ]
}`

and then write it out to the file system. So we'll write a function or two to take care of that.

```python
import time

def get_ocr_output_for_sequence(sequence_url=None):
    url = sequence_url[:-5] + "/ocr.txt" #this would be brittle to changing file types but works nicely for now
    rsp = requests.get(url)
    if rsp.status_code != 200:
        raise Exception(f"Unable to load {url}")
    ocr_output = rsp.text
    return ocr_output


def get_ocr_output_for_list_of_issues(list_of_issues=[]):
    issues_with_ocr = []
    
    for issue in list_of_issues:
        issue_data = get_issue(issue['url'])
        
        source = issue_data['url']
        title = f"{issue_data['title']['name']} {issue_data['date_issued']}"
        part_of = issue_data['title']['name']
        volume = issue_data['volume']
        number = issue_data['number']
        date_issued = issue_data['date_issued']
        
        pages = []
        
        for page in issue_data['pages']:
            time.sleep(0.05) #still worried about the API limits since they're not very forthcoming about it.
            ocr_output = get_ocr_output_for_sequence(sequence_url=page['url'])
            pages.append(ocr_output)
            
        issue_struct = {
            "source": source,
            "title": title,
            "isPartOf": part_of,
            "datePublished": date_issued,
            "volume": volume,
            "number": number,
            "text": pages
        }
            
        issues_with_ocr.append(issue_struct)
    
    return issues_with_ocr
        
ocr_issues_paper_0 = get_ocr_output_for_list_of_issues(paper_0_issues_to_retrieve)
ocr_issues_paper_1 = get_ocr_output_for_list_of_issues(paper_1_issues_to_retrieve)
```

```python
ocr_issues = ocr_issues_paper_0 + ocr_issues_paper_1
```

```python
import json

data = {"issues": ocr_issues}

with open('issues.json', 'w') as outfile:
    json.dump(data, outfile)
```
